```python
data = pd.read_csv("../data(S1_plastic).csv", index_col=False)
data
```

```python
a_ratio = []
```

```python
data.values[0]
```

```python
for i in range(len(data.values)):
    image_width = data.values[i][3]
    image_heigth = data.values[i][4]
    x1 = data.values[i][5]
    y1 = data.values[i][6]
    x2 = data.values[i][7]
    y2 = data.values[i][8]
    image_size = image_width*image_heigth
    bbox_size = (x2-x1)*(y2-y1)
    a_ratio.append(np.round(bbox_size/image_size*100,2))
```

```python
data['ascept_ratio'] = a_ratio
```

```python
data[data['ascept_ratio']==55.100000]
```
