```python
import pandas as pd # pandas 모듈 로드
import json  # json 모듈 로드
import os #os 모듈 로드
```

```python
#[path] 에 데이터가 있는 디렉터리
path = 'E:/reusable-trash-images/selectstar-reusable-trash-json' 

# 경로 내에 파일을 모두 불러옴
file_list = os.listdir(path)

#json 확장자를 가진 파일로 새로운 리스트 생성
json_file_list = [path+'/'+ file for file in file_list if file.endswith('.json')] 
# Python List Comprehesion사용
# 1. file_list라는 전체 파일을 불러놓은 변수에서 file을 값을 하나씩 빼서 사용
# 2. 조건문 if로 파일끝나는 부분이 .json확장자를 가질 때 file로 추출가능!
# 3. json_file_list에 경로 + / + file을 적으면 해당 파일의 전체 경로로 값을 바꿔서 저장한다!

json_file_list
```

```python
imagePath = [] #이미지 파일명 저장 리스트
label=[] #분류 라벨 값 저장 리스트
points_1=[] #boundary box x,y좌표 리스트
points_2=[] #boundary box x,y좌표 리스트
errorFile =[] #에러 파일 검출 리스트
```

```python
shapes_list = []
```

```python
for i in json_file_list: #json_file_list의 하나의 인덱스 마다 i라는 변수로 사용한다.
    with open(i,'r') as j: #
        try:
            contents=json.loads(j.read())
            shapes_list.append(len(contents['shapes']))
            imagePath.append((contents['imagePath']).replace('.jpg',""))
        except:
            errorFile.append(i) 
```

```python
t_df = pd.DataFrame({
    "imagePath":imagePath,
    "shapes":shapes_list
})
```
