```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
data = pd.read_csv("./first_df.csv", index_col=False)
data
```

```python
def ratio_cal(dx,dy,size):
    area = (dx*dy)/size
    x = int(np.sqrt(area*(dx/dy)).round())
    y = int(((dy/dx)*x).round())
    return x,y
```

```python
def make_points(array):
    x1 = array[5]
    y1 = array[6]
    x2 = array[7]
    y2 = array[8]
    dx = np.abs(x2-x1)
    dy = np.abs(y2-y1)
    image = str(array[0])
    return x1,y1,x2,y2,dx,dy,image
```

```python
error_file = []
```

```python
imagePath = "E:/reusable-trash-images/selectstar-reusable-trash-image/"

for i in range(len(data.values)):
    try:
        x1,y1,x2,y2,dx,dy,image = make_points(data.values[i])
        dx2, dy2 =ratio_cal(dx,dy,0.25)
        delta_x =int((dx2-dx)/2)
        delta_y =int((dy2-dy)/2)
        pt1_x = x1-delta_x
        pt1_y = y1-delta_y
        pt2_x = x2+delta_x
        pt2_y = y2+delta_y
        if pt1_x <0:
            pt1_x = x1-int(delta_x/2)
            pt2_x = x2+int(delta_x/2)
        if pt1_y <0:
            pt1_y = y1-int(delta_y/2)
            pt2_y = y2+int(delta_y/2)
        path = imagePath + image +".jpg"
        img = cv2.imread(path, cv2.IMREAD_COLOR)
        roi = img[pt1_y:pt2_y,pt1_x:pt2_x]
        print((dx*dy)/(roi.shape[0]*roi.shape[1]))
        if ((dx*dy)/(roi.shape[0]*roi.shape[1])>=1):
            error_file.append(image)
        else:
            cv2.imwrite(f"./data(75)/{image}-75.jpg", roi)
    except:
        error_file.append(image)
```
